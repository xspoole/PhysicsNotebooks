### PH 424 HW 2b

```python
# charger

def charger(t):
    return 36*10**(-9)*np.e**(-1136*t)*np.cos(50239*t)

t_vals = np.linspace(0, 1, 100000)
y_vals = charger(t_vals)
```

```python
plt.plot(t_vals, y_vals)
plt.xlim(0,0.001)
plt.ylim(-36*10**(-9), 36*10**(-9))
plt.ylabel('Charge (C)')
plt.xlabel('Time (s)')
plt.title('Charge through the LRC Circuit')
plt.grid()
plt.show
plt.savefig("PH424HW/HW1_3g_ChargePlot.png")
```

```python
def current(t):
    return -36*10**(-9)*np.e**(-1136*t)*50239*np.sin(50239*t)

t_vals = np.linspace(0, 1, 100000)
y_vals = current(t_vals)
```

```python
plt.plot(t_vals, y_vals)
plt.xlim(0,0.001)
plt.ylim(-36*10**(-4), 36*10**(-4))
plt.ylabel('Current (A)')
plt.xlabel('Time (s)')
plt.title('Current through the LRC Circuit')
plt.grid()
plt.show
plt.savefig("PH424HW/HW1_3g_CurrentPlot.png")
```

```python
def currentder(t):
    return -36*10**(-9)*np.e**(-1136*t)*50239**2*np.cos(50239*t)

t_vals = np.linspace(0, 1, 100000)
y_vals = currentder(t_vals)
```

```python
plt.plot(t_vals, y_vals)
plt.xlim(0,0.001)
plt.ylim(-90, 90)
plt.ylabel(r'$\frac{dI}{dt} (\frac{A}{s})$')
plt.xlabel('Time (s)')
plt.title('Derivative of Current through the LRC Circuit')
plt.grid()
plt.show
plt.savefig("PH424HW/HW1_3g_CurrentDerivPlot.png")
```

```python
# Graph 1 for the annoying potentiometer problem ew

def ewdiode(I):
    return 1100*I + 0.025*np.log(I/(10**(-9)))
```

```python
I = np.linspace(1e-9, 5e-3, 500)
V = ewdiode(I)

# Target line
y_line = 5

# Find approximate intersection
# (find where the difference changes sign)
idx = np.argwhere(np.diff(np.sign(V - y_line))).flatten()

if len(idx) > 0:
    I_int = I[idx[0]]
    V_int = V[idx[0]]
else:
    I_int, V_int = None, None

# Plot
plt.figure(figsize=(8,5))
plt.plot(I, V, label=r'$V_s = 1100I + 0.025\ln{\left(\frac{I}{10^{-9}}\right)}$', color = 'blue')
plt.axhline(y_line, color='r', linestyle='--', label=r'$V_s = 5$')

# Label intersection point if found
if I_int is not None:
    plt.scatter(I_int, V_int, color='black', zorder=5)
    plt.text(I_int*0.6, V_int + 0.2, 
             fr'Intersection: $I = 0.0042$ A', 
             fontsize=10, color='black')

plt.title('Determining Current at the Ammeter')
plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')
plt.grid(True)
plt.legend(loc='lower right')
plt.savefig("ph411graphs/problemset2q1.png")
```

```python
# Graph for Q.3

# Define the function
def Vc(t):
    return 5 * (1 - np.exp(-t / (10**-4)))  # 5(1 - e^(-t/10^-4))

# Define a time range
t = np.linspace(0, 5e-4, 500)  # from 0 to 0.0005 seconds (5)
V = Vc(t)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, V, label=r'$V_c(t) = 5(1 - e^{\frac{-t}{10^{-4}}})$', color='blue')

# Labels, grid, legend
plt.title('Voltage of a Capacitor vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.grid(True)
plt.legend(loc='lower right')
plt.savefig("ph411graphs/problemset2q3.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
T = 1.0   # choose any T you want

# Frequency range
omega = np.linspace(-50, 50, 2000)

# Compute the sinc argument
x = (omega * T) / 2

# numpy's sinc is sin(pi x)/(pi x), so we need sin(x)/x:
sinc = np.sinc(x / np.pi)

# Full expression
H = (T / (2 * np.pi)) * np.exp(-1j * omega * T / 2) * sinc

# Plot
plt.figure(figsize=(10,6))
plt.plot(omega, H.real, label='Real Part')
plt.plot(omega, H.imag, label='Imag Part', linestyle='--')
plt.xlabel(r'$\omega$')
plt.ylabel('Magnitude')
plt.title(r'$\frac{T}{2\pi} e^{-i\omega T/2} \, \mathrm{sinc}(\omega T/2)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def cot(x):
    return np.cos(x)/np.sin(x)

def ddottheta(x):
    return np.sin(x)*(np.cos(x) - 10)

def ddotphi(x):
    return -2*cot(x)

np.linspace(1, 100, 10000)
```

```python
plt.plot(x_vals, ddottheta(x_vals), linewidth = 2)
plt.plot(x_vals, ddotphi(x_vals), linewidth = 2)

plt.title(r'Angular Acceleration $\ddot{\theta}$ and $\ddot{\phi}$ for Varied $\theta$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\ddot{\theta}, \ddot{\phi}$')
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True, alpha = 0.5)

plt.savefig('ph411graphs/phiandtheta.png')
```

```python
def cot_deg(x):
    radians = np.radians(x)
    return np.cos(radians) / np.sin(radians)

def ddottheta_deg(x):
    radians = np.radians(x)
    return np.sin(radians) * (np.cos(radians) - 10)

def ddotphi_deg(x):
    return -2 * cot_deg(x)

# Generate x-values (in degrees)
x_vals = np.linspace(1, 360, 10000)

# Compute values
y1 = ddottheta_deg(x_vals)
y2 = ddotphi_deg(x_vals)

# Mask infinite or invalid values to avoid plot breaks
y1 = np.where(np.isfinite(y1), y1, np.nan)
y2 = np.where(np.isfinite(y2), y2, np.nan)

# Plot
plt.plot(x_vals, y1, linewidth=2, label=r'$\ddot{\theta}$')
plt.plot(x_vals, y2, linewidth=2, label=r'$\ddot{\phi}$')

plt.title(r'Angular Accelerations $\ddot{\theta}$ and $\ddot{\phi}$ for Varied $\theta$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\ddot{\theta}, \ddot{\phi}$')
plt.xlim(0, 360)
plt.ylim(-100, 100)
plt.legend()
plt.grid(True, alpha=0.5)

plt.savefig('ph411graphs/phiandtheta.png')
```
